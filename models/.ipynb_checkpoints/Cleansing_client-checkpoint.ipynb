# I) Data Understanding

```python
#Importings
import pymysql
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime as dt
import matplotlib.pyplot as plt
import warnings
```

```python
#Connect to mysql
conn=pymysql.connect(host='localhost',port=int(3308),user='root',passwd='',db='fidness')
```

```python
df=pd.read_sql_query("SELECT c.id, c.civilite, c.nom_prenom, c.email, c.date_naissance, c.date_inscription, g.nom\
       FROM fid_client c \
       INNER JOIN fid_gouvernorat as g ON c.id_gouvernorat = g.id",conn)
df.rename(columns = {'nom':'pays'}, inplace = True)
df
```

# II) Data Cleaning, Preprocessing and Visualization

```python
#verifiy null values
df.isnull().sum()
```

```python
#pourcentage of null values
df.isnull().sum() / len(df)*100
```

```python
#drop null values
df=df.dropna(axis=0)
df.isna().sum()
```

```python
#checking duplicates
df.duplicated().sum()
```

```python
#creation of gender col
warnings.filterwarnings("ignore", category=DeprecationWarning) 
warnings.filterwarnings("ignore", category=FutureWarning) 
warnings.filterwarnings("ignore", category=UserWarning)
l=[]
for item in df['civilite']:
    if item == 1:
        l.append("homme")
    else:
        l.append("femme")
df["sexe"] = l
df
```

```python
#convert datetime to date
warnings.filterwarnings("ignore", category=DeprecationWarning) 
warnings.filterwarnings("ignore", category=FutureWarning) 
warnings.filterwarnings("ignore", category=UserWarning)
from datetime import datetime, date
for date_n, date_i in zip(df['date_naissance'],df['date_inscription']):
    df['date_naissance'] = pd.to_datetime(df['date_naissance']).dt.date
    df['date_inscription']=pd.to_datetime(df['date_inscription']).dt.date
```

```python
#Calculate age
warnings.filterwarnings("ignore", category=DeprecationWarning) 
warnings.filterwarnings("ignore", category=FutureWarning) 
warnings.filterwarnings("ignore", category=UserWarning)
l1=[]
for date in df['date_naissance']:
    l1.append(date)
df['age']=pd.to_datetime('today').year-pd.to_datetime(l1).year
```

```python
#select user table
df1=pd.read_sql_query("SELECT * FROM fid_client_palier_tampon ",conn)
df1
```

```python
#checking if customer has tampon card or not
l6=[]
for id in df['id']:
    if id in df1['id_client']:
        l6.append(1)
    else:
        l6.append(0)
df['a_une_carte'] = l6
```

```python
x=df1.groupby(['id_client']).sum()
x
```

```python
#appending total tampon number to client id
l7=[]
for id in df['id']:
    if id in x.index:
        l7.append(x['nbre_tampon_actuelle'][id])
    else:
        l7.append(0)
df['nbre_total_tampon'] = l7
```

```python
#Calculate number of years that the customer has been a client of fidness
warnings.filterwarnings("ignore", category=DeprecationWarning) 
warnings.filterwarnings("ignore", category=FutureWarning) 
warnings.filterwarnings("ignore", category=UserWarning)
l8=[]
for date in df['date_inscription']:
    l8.append(date)
df['nombre_annees']=pd.to_datetime('today').year-pd.to_datetime(l8).year
```

```python
#verify if customer is active or not
l2=[]
l3=[]


df1["Active"]=0

for i,j in zip(df1["id_client"],df1["date_achat"]):
        date = pd.to_datetime(j).strftime('%Y/%m/%d')
        res = (dt.strptime('2022/6/1', "%Y/%m/%d") - dt.strptime(date, "%Y/%m/%d")).days
        l2.append(res)
     
                
        
for i in l2:
    if i<30*12+5:
        l3.append(1)
    else:
        l3.append(0)
            
        
df1["Active"]=l3
```

```python
#assigning quitted and non quitted customers
warnings.filterwarnings("ignore", category=DeprecationWarning) 
warnings.filterwarnings("ignore", category=FutureWarning) 
warnings.filterwarnings("ignore", category=UserWarning)
df['quitte'] = 0
for i in s2:
    for j in range(10008):
        if i==df["id"][j]:
            df["quitte"][j]=1
```
