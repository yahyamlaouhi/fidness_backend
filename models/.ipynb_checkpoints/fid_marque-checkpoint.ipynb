```python
#Connect to mysql
import pymysql
import pandas as pd
import numpy as np
conn=pymysql.connect(host='localhost',port=int(3308),user='root',passwd='',db='fidness')
```

```python
#select user table
df=pd.read_sql_query("SELECT * FROM fid_marque ",conn)
df.tail()
```

```python
#checking null values
df.isnull().sum()
```

```python
#exportiiiing
from sqlalchemy import create_engine
engine = create_engine('mysql+pymysql://root:@localhost:3308/fid_db')
df.to_sql(name='fid_marque',con=engine,if_exists='replace', index=False)
```
