```python
#Connect to mysql
import pymysql
import pandas as pd
import numpy as np
conn=pymysql.connect(host='localhost',port=int(3308),user='root',passwd='',db='fidness')
```

```python
#select user table
df=pd.read_sql_query("SELECT * FROM fid_client ",conn)
df.head()
```

```python
#drop uneeded columns
df.drop(df.columns[[3,6,7,8,9,10,11,12,13,5]], axis=1, inplace=True)
df.head()
```

```python
#exportiiiing
from sqlalchemy import create_engine
engine = create_engine('mysql+pymysql://root:@localhost:3308/fidness')
df.to_sql(name='fid_client',con=engine,if_exists='replace', index=False)
```
